```
train_df = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    "amananandrai/ag-news-classification-dataset",
    "train.csv"
)

print("Train shape:", train_df.shape)
train_df.head()
```

```
test_df = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    "amananandrai/ag-news-classification-dataset",
    "test.csv"
)

print("Test shape:", test_df.shape)
test_df.head()
```

```
train_df.columns = ["label", "title", "description"]
test_df.columns = ["label", "title", "description"]

train_df["text"] = train_df["title"] + " " + train_df["description"]
test_df["text"] = test_df["title"] + " " + test_df["description"]

train_df["label"] -= 1
test_df["label"] -= 1

train_df.head()
```

```
import re
import string

def clean_text(text):
    text = text.lower()
    text = re.sub(r"\d+", "", text)
    text = text.translate(str.maketrans("", "", string.punctuation))
    text = re.sub(r"\s+", " ", text).strip()
    return text

train_df["clean_text"] = train_df["text"].apply(clean_text)
test_df["clean_text"] = test_df["text"].apply(clean_text)
```

```
from sklearn.feature_extraction.text import CountVectorizer

bow = CountVectorizer(max_features=10000)

X_train_bow = bow.fit_transform(train_df["clean_text"])
X_test_bow = bow.transform(test_df["clean_text"])
```

```
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=10000)

X_train_tfidf = tfidf.fit_transform(train_df["clean_text"])
X_test_tfidf = tfidf.transform(test_df["clean_text"])

y_train = train_df["label"]
y_test = test_df["label"]
```

```
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report

linear_svm = LinearSVC(C=1)
linear_svm.fit(X_train_tfidf, y_train)

y_pred_linear = linear_svm.predict(X_test_tfidf)

print("Linear SVM Results")
print(classification_report(y_test, y_pred_linear))
```

```
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=10000)

X_train_tfidf = tfidf.fit_transform(train_df["clean_text"])
X_test_tfidf  = tfidf.transform(test_df["clean_text"])

y_train = train_df["label"]
y_test  = test_df["label"]

print("TF-IDF shapes:", X_train_tfidf.shape, X_test_tfidf.shape)
```

```
subset_size = 15000

X_train_small = X_train_tfidf[:subset_size]
y_train_small = y_train[:subset_size]

print("Subset shape:", X_train_small.shape)
```

```
from sklearn.svm import SVC
from sklearn.metrics import classification_report

rbf_svm = SVC(kernel="rbf", C=1, gamma="scale")

rbf_svm.fit(X_train_small, y_train_small)

y_pred_rbf = rbf_svm.predict(X_test_tfidf)

print("RBF SVM Results (Subset)")
print(classification_report(y_test, y_pred_rbf))
```

```
from sklearn.svm import LinearSVC
from sklearn.model_selection import GridSearchCV

param_grid = {
    "C": [0.01, 0.1, 1, 10]
}

grid = GridSearchCV(
    LinearSVC(),
    param_grid,
    cv=3,
    scoring="f1_macro"
)

grid.fit(X_train_tfidf, y_train)

print("Best Linear SVM Params:", grid.best_params_)
best_svm = grid.best_estimator_
```

```
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report

# Predict using optimized Linear SVM
y_pred_final = best_svm.predict(X_test_tfidf)

print("Final Optimized Linear SVM Performance")
print("Accuracy :", accuracy_score(y_test, y_pred_final))
print("Precision:", precision_score(y_test, y_pred_final, average="macro"))
print("Recall   :", recall_score(y_test, y_pred_final, average="macro"))
print("F1 Score :", f1_score(y_test, y_pred_final, average="macro"))

print("\nDetailed Classification Report:")
print(classification_report(y_test, y_pred_final))
```

```
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report

nb = MultinomialNB()

nb.fit(X_train_tfidf, y_train)
y_pred_nb = nb.predict(X_test_tfidf)

print("Naive Bayes Performance")
print(classification_report(y_test, y_pred_nb))

```
